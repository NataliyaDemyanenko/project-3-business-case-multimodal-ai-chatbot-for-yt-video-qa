```python
import yt_dlp
from youtube_transcript_api import YouTubeTranscriptApi
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
from tqdm.auto import tqdm  # Import tqdm for progress bar

def preprocess_video(video_id):
    # 1. Download video metadata
    ydl_opts = {'skip_download': True}
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(f"https://www.youtube.com/watch?v={video_id}", download=False)

    # 2. Get transcript
    transcript = YouTubeTranscriptApi.get_transcript(video_id)

    # 3. Split text into chunks
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=200)
    chunks = []
    metadatas = []
    ids = []

    total_chunks = 0

    # Calculate total number of chunks for tqdm
    for entry in transcript:
        chunk = text_splitter.split_text(entry['text'])
        total_chunks += len(chunk)

    # Progress bar setup
    with tqdm(total=total_chunks, desc=f"Processing video {video_id}", unit="chunk") as pbar:
        for i, entry in enumerate(transcript):
            chunk_texts = text_splitter.split_text(entry['text'])
            for j, text in enumerate(chunk_texts):
                chunk_id = f"{video_id}_{i}_{j}"
                video_data[chunk_id] = {
                    "video_id": video_id,
                    "title": info['title'],
                    "channel": info['channel'],
                    "description": info['description'],
                    "duration": info['duration'],
                    "start_time": entry['start'],
                    "end_time": entry['start'] + entry['duration'],
                    "source": f"https://www.youtube.com/watch?v={video_id}&t={int(entry['start'])}",
                    "chunk_text": text
                }
                pbar.update(1)  # Update progress bar for each processed chunk

    return video_data
```

```python
video_ids = ["wKr8jECz53U"]
for video_id in video_ids:
  video_data= {}
  preprocess_video(video_id)
```
